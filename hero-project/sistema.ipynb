## Introdução
O <span class="game-title">Hero Project</span> é um jogo sobre super-heróis com um sistema de jogo de interpretação minimalista e focado na narrativa. O <abbr title="Mestre do Jogo">*Game Master*</abbr><sup>1</sup> (GM) cria o mundo e controla os <abbr title="Personagens Não-Jogadores">*Non-Player Characters*</abbr> (NPCs). Todos os demais controlam um <abbr title="Personagens Jogadores">*Player Character*</abbr> (PC), narrando as ações de seu protagonista conforme ele conduz a história e interage com o mundo do jogo.

<div class="footnote"><p><sup>1</sup> Posicione o cursor do mouse em textos com <abbr title="e você verá informações adicionais, como traduções, explicações sobre termos, etc.">essa formatação</abbr>.</p></div>

### Visão Geral do Sistema
Os jogadores rolam de 1-3 dados de seis lados (`3d6`) (dependendo do **Arquétipo** de seu personagem) contra uma **Dificuldade** de 4-6 (dependendo da situação) e precisam ser bem-sucedidos com pelo menos um dado.

**Regalias** podem reduzir a Dificuldade, enquanto **Peculiaridades** podem aumentá-la. Os jogadores gastam **Carma** nas Regalias e podem recuperar carma a partir das Peculiaridades.

Os jogadores realizam *todas as rolagens*. O GM *nunca* rola os dados – ele apenas descreve os **Desafios**, atribui os **Traços** e a Dificuldade e narra o resultado.

### Exemplo de Jogo
Se você estiver lendo este exemplo pela primeira vez, provavelmente não entenderá tudo o que está acontecendo. Mas não se preocupe, tudo fará sentido em breve.

## Personagens
### Criação
Para criar um personagem, escolhe um nome, um Arquétipo, uma Regalia e uma Peculiaridade. Você pode escrever seu personagem como uma frase descritiva, como "um cavalheiro bruto de criação nobre com um código de honra estrito" ou "um jornalista ágil com superpoderes de aranha e uma tendência de fazer piadas."

#### Arquétipo
O Arquétipo de um personagem consiste de um **Conceito** (frequentemente uma profissão, mas também pode incluir uma raça ou outros descritores) combinado com um Traço (ágil, astuto ou bruto).

Se seu conceito de personagem possuir poderes mágicos, você será capaz de superar muitos Desafios com magia – mas se você também não possuir uma Regalia apropriada, você não será capaz de gastar Carma em seus feitiços.

**Exemplos:** Assassino ágil, patrulheiro elfo ágil, soldado bruto, paladino bruto, jornalista astuto, feiticeiro astuto.

Leia mais em [Arquétipos](#Arquétipos).

#### Regalias
As Regalias representam talentos especiais, habilidades, perícias, itens arcanos, profissões, etc. Você inicia com uma Regalia de sua escolha.

**Exemplos:** <span class="perk">Charmoso</span>, <span class="perk">Espada Mágica</span>, <span class="perk">Necromancia</span>, <span class="perk">Linhagem Nobre</span>, <span class="perk">Psiônico</span>, <span class="perk">Reflexos Rápidos</span>, <span class="perk">Estudioso do Oculto</span>.

Leia mais em [Regalias](#Regalias).

#### Peculiaridades
As Peculiaridades representam obstáculos, deficiências e traços de personalidade negativos que podem dificultar a vida para o personagem. Você inicia com uma Peculiaridade de sua escolha.

**Exemplos:** <span class="quirk">Arrogante</span>, <span class="quirk">Visão Ruim</span>, <span class="quirk">Maldoso</span>, <span class="quirk">Perna de Pau</span>, <span class="quirk">Impiedoso</span>, <span class="quirk">Feio</span>, <span class="quirk">Vingativo</span>.

Leia mais em [Peculiaridades](#Peculiaridades).

#### Carma & Determinação
O Carma representa sorte e providência, enquanto a **Determinação** representa tenacidade, espírito, saúde e resolução.

Os jogadores iniciam cada sessão com três pontos de Carma e três pontos de Determinação. Esses pontos podem ser recuperados durante o jogo, mas nunca podem exceder a quota inicial do personagem na sessão.

#### Equipamentos
Os **Equipamentos** são uma ferramenta narrativa usada para justificar as capacidades de um personagem ao invés de fornecerem bônus. Os jogadores podem descrever seus Equipamentos para os próprios descritivos, mas assume-se que geralmente eles possuem o que for necessário para seu Arquétipo e Regalias.

Se você quiser que seu personagem possua um item excepcional ou arcana que fornece benefícios significativos, selecione-o como uma Regalia.

### Avanço
Os personagens geralmente avançam a cada 1-3 sessões, conforme critério do GM, baseado na duração da aventura.

Quando seu personagem avança, adicione uma nova Regalia ou Peculiaridade de sua escolha — ou converta uma **Aflição** em uma Peculiaridade! Isso representa uma habilidade, fraqueza ou item que seu personagem encontrou ou desenvolveu durante a história.

A cada dois avanços você pode aumentar seu Carma *ou* Determinação em um ponto (até um máximo de seis cada) *ao invés* de adicionar uma Regalia ou Peculiaridade.

## Regras do Jogo
### Desafios
O jogo envolve superar vários Desafios, cada um com uma Dificuldade de 4-6 (fácil, padrão ou difícil). A maioria terá Dificuldade 5, mas o GM pode decidir que alguns Desafios são mais fáceis ou mais difíceis.

Os jogadores rolam de `1-3d6` dependendo de seu Arquétipo. Se um dado igualar ou exceder a Dificuldade, o personagem é bem-sucedido – caso contrário, ele fracassa. Se igualar ou exceder a Dificuldade em 2-3 dados, é um sucesso *excepcional*.

Se o jogador rolar `1` em todos os seus dados, o resultado é uma fracasso *crítico* e isso é sempre *muito* ruim!

Alguns Desafios requerem esforço adicional para serem superados. Isso é representado pelo uso de pontos de **Esforço**; cada dado que igual ou vencer a Dificuldade remove um ponto e o Desafio é derrotado uma vez que todos os pontos tenham sido removidos. Os PCs geralmente podem trabalhar juntos para superar tal Desafio e serão necessárias várias rolagens; falhar em qualquer uma dessas rolagens terá consequências.

#### Fracasso & Sucesso
Um Desafio possui quatro resultados possíveis: *sucesso exceptional*, *sucesso normal*, *fracasso normal* e *fracasso crítico*.

Esses resultados são relativos ao personagem e à situação, conforme decidido pelo GM, e nenhum resultado, seja no melhor ou no pior dos casos, será tão extremo a ponto de exigir a suspensão da descrença. Um ladrão mestre nunca *falharia* de maneira crível na abertura de uma fechadura simples, embora possa levar mais tempo que o necessário ou causar uma **Complicação**. Da mesma forma, um estudioso desarmado nunca poderia derrotar uma dezena de guerreiros de elite em combate corporal – no melhor dos casos, ele conseguir fugir intacto.

Se alguém estiver tentando traduzir um texto mágico, e não possuir nenhum antecedente em magia ou no idioma usado, até mesmo um *sucesso excepcional* forneceria menos informações para ele do que um fracasso normal forneceria para um mago estudioso!

Contudo, sempre haverá algum preço para o fracasso. Às vezes esse preço é óbvio, como ser visto enquanto tenta passar despercebido. Outras vezes o GM pode remover um ponto de Determinação de um personagem ou introduzir uma Complicação na cena.

Um fracasso crítico é sempre muito ruim, não importa o quão habilidoso o personagem seja e frequentemente representa um golpe de azar. Se o personagem perderia um ponto de Determinação em um fracasso normal, ele perderia dois em um fracasso crítico.

#### Derrota & Aflições
Os personagens são derrotados quando ficarem sem Determinação e o vitorioso (jogador ou GM) decide o destino da vítima.

Personagens derrotados adquirem uma Aflição apropriada à situação, como um <span class="affliction">Braço Quebrado</span>, uma <span class="affliction">Fobia</span>, um <span class="affliction">Ego Machucado</span>, etc. Eles recuperam toda a Determinação, mas geralmente ficam incapazes de participar ativamente pelo restante da cena – eles podem estar inconscientes no chão, fugindo em pânico ou apenas muito machucados para continuar.

As Aflições são descritas pelo vitorioso e tratadas como Peculiaridades temporárias (ou permanentes se forem causadas por um fracasso crítico), exceto que o GM decide quando aplicá-las. Um PC com mais de três Aflições é removido do jogo, embora possam retornar se uma ou mais de suas Aflições forem curadas.

Morte é uma questão narrativa. Embora um jogador possa decidir matar seu adversário em combate conferindo-lhe uma Aflição fatal, o GM sempre alertará os jogadores quando um Desafio puder resultar em morte.

#### Recuperação
Certas Aflições (e.g. <span class="affliction">Fugindo em Pânico</span>) são removidas automaticamente no final da cena, mas outras podem durar horas, dias ou mais, conforme critério do GM. Um PC com uma Regalia adequada pode gastar Carma para curar uma Aflição (e.g. <span class="perk">Regeneração</span> para curar um ferimento) – mas Aflições *permanentes* custam Carma *permanentemente* (exceto se forem convertidas em Peculiaridades através do avanço).

#### Desafios Opostos
A maioria dos Desafios consiste nos PCs trabalhando juntos contra ameaças externas, mas ocasionalmente, eles podem desejar se opor *entre si*. Se isso ocorrer, ambos os jogadores rolam normalmente, mas cada um tratará o dado mais alto na rolagem do outro como a Dificuldade de seu *próprio* Desafio, em outras palavras, a rolagem mais alta vence.

Em um empate, o jogador que equiparou a Dificuldade com o maior número de dados alcança um sucesso normal (e.g. `4, 4, 4` é um sucesso normal contra `4, 4, 3` ou `4, 3, 3`). Caso o empate permaneça, o GM tentará interpretar o resultado de uma maneira que favoreça ambos os lados igualmente.

Se ambos os jogadores rolarem um fracasso crítico, então cada um sofrerá um resultado terrível.

#### Confrontos entre NPCs
Existem muitas situações onde dois NPCs possuem um confronto direto entre si, ao invés de contra os PCs. O GM geralmente decide o resultado, mas pode solicitar aos jogadores que rolem pelos NPCs, caso queira um resultado imprevisível.

### Combate
O combate e outros conflitos são resolvidos em trocas de ataques em turnos.

#### NPCs como Desafios
O GM atribui adversários com uma Dificuldade de 4-6, baseado em seu poder relativo ao PC. A maioria dos inimigos terá Dificuldade 5.

Cada adversário também possui um ou mais pontos de Esforço que representam sua Determinação. O GM pode agrupar um grupo de inimigos semelhantes, como uma horda de goblins. Nesse caso, eles são tratados como um único Desafio com pontos de Esforço adicionais.

#### Resolução
A ordem dos turnos seguirá a narrativa sempre que possível e os jogadores realizam **todas as rolagens** – eles rolam para atacar em seu turno e para defender no turno do inimigo.

Como regra geral, os jogadores só podem realizar uma rolagem de defesa por turno. Se estiverem enfrentando vários adversários, devem rolar contra o atacante mais perigoso.

### Arquétipos
O Arquétipo de um personagem consiste um Conceito (geralmente uma profissão) combinado com um Traço (ágil, astuto ou bruto) e essa combinação determina quantos dados ele terá para rolar em Desafios.

Personagens **ágeis** rolam `3d6` para qualquer coisa relacionada com rapidez, destreza, reflexos ou furtividade. Eles também rolam `3d6` para combate de longo alcance (mas veja mais sobre isso adiante).

Personagens **astutos** rolam `3d6` quando realizam Desafios relacionados com carisma, intelecto, força de vontade ou percepção. Eles também rolam `3d6` para combate mental (mas veja mais sobre isso adiante).

Personagens **brutos** rolagem `3d6` para quaisquer Desafios baseados em força, resistência física, vigor ou atletismo. Eles também rolam `3d6` para combate corporal (mas veja mais sobre isso adiante).

Quando os personagens carecem do Traço apropriado para um Desafio, eles rolam apenas `2d6`. Se um determinado Desafio necessitar de Conhecimento Especial que caia fora do escopo de seu Conceito e Regalias, então o número de dados rolados é reduzido em 1.

#### Exemplos de Arquétipos
Um bárbaro bruto rola `3d6` para empunhar uma espada, `2d6` para arremessar uma lança e `1d6` para arrombar uma fechadura. Um patrulheiro elfo ágil rola `3d6` para esgueirar-se silenciosamente pela floresta, `2d6` para localizar um inimigo escondido e `1d6` para negociar um tratado legal. Um mago astuto rola `3d6` para atirar uma bola de fogo em alguém, `2d6` para escalar uma corda e `1d6` para empunhar uma espada grande em combate.

#### Estilos de Combate
O estilo de combate preferido de um personagem (corporal, longo alcance ou mental) *geralmente* é baseado em seu Traço. Contudo, isso pode ser alterado se outro estilo se adequar melhor ao conceito do personagem. Por exemplo, um ladrão ágil pode preferir armas corporais e um atirador astuto muito provavelmente usaria armas de longo alcance.

O estilo de combate do personagem deve ser escolhido durante a criação do personagem e não pode ser alterado posteriormente.

**Observação:** Combate mental inclui ataques mágicos, persuasão, intimidação, etc.

#### O Contexto é Importante
Personagens astutos geralmente rolam `3d6` para Desafios sociais, mas isso não significa que eles sempre são melhores nisso.

Uma bruxa velha astuta normalmente rola `3d6`, mas cortejar um homem jovem foge do escopo de seu conceito, portanto ela rolaria apenas `2d6` em tal situação. Talvez ela também tenha uma Peculiaridade <span class="quick">Nariz Verrugoso</span> – se sim, isso também aumentaria a Dificuldade do Desafio ou adicionaria uma Complicação.

Um sedutor ágil normalmente rola `2d6`, mas cortejam uma dama jovem está dentro do escopo de seu conceito, então ele não perderia um dado – e ele poderia possuir uma Regalia <span class="perk">Charmoso</span>, que reduziria a Dificuldade do Desafio. Ele poderia até mesmo gastar um ponto de Carma para descrever a jovem se encantando por ele.

Mas, obviamente, o sedutor rolaria `1d6` para persuadir um marido rejeitado à lhe dar um tempo e a bruxa velha astuta rolaria `3d6` para convencer uma mulher jovem a experimentar sua roda de fiar ou dar uma mordida em uma suculenta maçã vermelha.

### Regalias
Os personagens com uma Regalia relevante podem solicitar dicas ou pistas para O GM sobre uma determinada situação, receber um benefício temporário (como um equipamento especial ou auxílio de um NPC), superar obstáculos que geralmente seriam impossíveis para outras pessoas (como usar um par de asas para voar até um telhado) e assim por diante.

Os benefícios e usos exatos de uma Regalia são sempre à critério do GM, mas uma Regalia definida amplamente possui um escopo maior e, portanto, uma Regalia definida mais estreitamente terá mais impacto quando entrar no jogo.

Se um jogador quiser usar sua Regalia para impactar a história de maneira *significativa*, ele deve gastar um ponto de Carma para tal.

O Carma também pode ser gasto para reduzir a Dificuldade de um Desafio em 1 – tome essa decisão *após* rolar e certifique-se de narrar como a Regalia lhe dá uma vantagem. Apenas um ponto de Carma pode ser gasto dessa forma para cada Desafio.

#### Uma Análise Mais Profunda
Se você quiser fazer alguma coisa que pessoas comuns não conseguem nem tentar, mas para a qual sua Regalia realmente deveria permitir que você fizesse automaticamente, você deve gastar um ponto de Carma. Por exemplo, usar <span class="perk">Necromancia</span> para conjurar e interrogar o espírito de uma vítima assassinada ou talvez <span class="perk">Intuição Investigativa</span> para colher informações abrangentes em um cena de assassinato.

Se você quiser fazer alguma coisa que normalmente não é possível, mas para a qual sua Regalia deveria permitir que você pelo menos *tentasse*, você não precisa gastar Carma, mas ainda deve rolar. Por exemplo, usar sua <span class="perk">Super-Força</span> para erguer um ônibus.

Se uma Regalia permite que você ignorar um Desafio automaticamente enquanto outras pessoas precisam rolar, você deve gastar um ponto de Carma. Por exemplo, usar <span class="perk">Voo</span> para voar por cima de um rio que todos os demais precisam nadar ou por ser <span class="perk">Podre de Rico</span> poder atirar um monte de dinheiro em um problema para resolvê-lo.

Se o Desafio geralmente é possível para todos, você ainda pode descrever como usa sua Regalia para aumentar as probabilidades e pode, posteriormente, gastar Carma retroativamente para reduzir a Dificuldade. Por exemplo, usar suas <span class="perk">Pernas de Canguru</span> para saltar até um telhado, enquanto todos os demais precisam escalar.

Se você usar uma Regalia apenas para adicionar descrições narrativas para a cena (e.g. superar um Desafio através de magia sendo que os demais poderiam facilmente superá-lo através de perícia), você não precisa gastar Carma. Por exemplo, você poderia atirar uma bola de fogo contra um adversário, enquanto outro personagem atira uma flecha; a diferença é apenas uma questão narrativa.

#### Outra Perspetiva
Sempre pensa nas Regalias em termos de superar Desafios. Você rola apenas quando é importante para a história, mas se puder usar uma Regalia para obter sucesso automático *sem* rolar, isso custa Carma. Da mesma forma, custa Carma reduzir a Dificuldade de um Desafio – mas se uma Regalia não oferece nenhum outro benefício, então *não* custa Carma.

Se um jogador não possuir Carma, o GM pode oferecer uma Complicação ao invés de cobrar Carma.

#### Conhecimento Relacionado
As Regalias geralmente implicam um grau de conhecimento em qualquer campo relacionado. Por exemplo, um <span class="perk">Mecânico Automotivo</span> teria algum grau de conhecimento geral sobre mecânica – ele pode não saber sobre aeronaves, por exemplo, mas ainda assim teria uma chance muito maior de consertar uma do que alguém com nenhuma habilidade mecânica. De modo semelhante, alguém com a Regalia <span class="perk">Mestre da Espada</span> poderia aplicar sua especialização marcial para outras situações de combate, uma Regalia <span class="perk">Cirurgião</span> também implica treinamento médico geral e assim por diante.

#### Auxiliando Aliados
Os jogadores também podem usar Regalias para auxiliar seus amigos e aliados. Por exemplo, um <span class="perk">Curandeiro Divino</span> poderia gastar Carma para remover uma Aflição de ferimento de um amigo ferido, enquanto um <span class="perk">Comandante Corajoso</span> poderia gastar Carma para inspirar um companheiro, reduzindo a Dificuldade de um Desafio baseado em medo.

Auxiliar alguém não altera o limite padrão de gasto de um Carma por Desafio; se você gastar Carma para auxiliar um aliado, ele não pode gastar seu próprio Carma para reduzir a Dificuldade ainda mais.

#### Empilhando Regalias
Embora não seja possível gastar mais de um Carma por Desafio, um PC ainda pode usar várias Regalias ao mesmo tempo – o sucesso e o fracasso são relativos às competências do personagem, portanto o GM levará em consideração todas as Regalias aplicáveis.

Da mesma forma, assim como o impacto de uma Regalia depende do quão estritamente definida ele é, várias Regalias relevantes devem ter mais impacto do que uma única Regalia.

### Peculiaridades
Os jogadores devem declarar sua intenção de usar uma Peculiaridade *antes* de rolarem para um Desafio. Eles devem descrevem as ações de seu personagem de uma forma que incorpore a Peculiaridade e então aumentar a Dificuldade do Desafio em 1 (isso pode levá-la acima de 6).

Os jogadores geralmente recuperar um ponto de Carma por usar sua Peculiaridade, mas se forem bem-sucedidos na rolagem do Desafio, ao invés disso, podem recuperar um ponto de Determinação (se desejarem). Apenas uma Peculiaridade pode ser usada em cada Desafio.

#### Complicações
O GM também pode oferecer pontos de Carma para os jogadores em troca de uma Complicação. Se o jogador aceitar essa oferta, o GM utiliza suas Peculiaridades como inspiração – o professor com <span class="quirk">Visão Ruim</span> pode ter deixado uma pista importante passar despercebida, enquanto o capanga <span class="quirk">Malvado</span> pode ter insultado a pessoa errada.

Obviamente, as Complicações também podem ser baseadas na situação ou até mesmo no Arquétipo ou Regalias. Mas, quando possível, o GM tentará associá-las com as Peculiaridades do personagem.

## Regras Específicas
### Cibernética
Assim como qualquer Equipamento mundano, itens cibernéticos serão tratados como detalhes narrativos de fundo ou como Regalias, dependendo de quanto impacto o item em questão terá dentro do jogo. Um indivíduo com aprimoramentos cibernéticos extensivos também deve possuir uma Peculiaridade adequada para representar os reveses físicos e psicológicos de seus vários implantes.

### Medo & Insanidade
O medo será tratado da mesma forma que outros Desafios: personagens astutos rolam `3d6`, enquanto os demais rolam `2d6` e indivíduos sem experiência ou exposição anterior à determinada fonte de medo (conforme seu Conceito e Regalias) reduzirão o número de dados rolados em 1. Fracasso em Desafios de medo resultam na perda de um ponto de Determinação – se o personagem perder toda a Determinação, ele geralmente foge da cena ou recebe uma forma apropriada de distúrbio mental.

### Adversários
#### Traços
Os adversários podem possuir um ou mais Traços: adversários ágeis, astutos e brutos aumentam a Dificuldade dos Desafios que utilizem esses Traços; adversários desajeitados, estúpidos ou fracos *reduzem* a Dificuldade dos Traços relacionados. Atirar contra um goblin "ágil e fraco" tem Dificuldade 6, por exemplo, enquanto acertá-lo em combate corporal tem Dificuldade 4.

#### Ranks
Os PCs iniciam no **Rank** 1 e sobem de Rank a cada 4 avanços (ou seja, no 4, 8, 12, 16 e 20) até um máximo de Rank 6. NPCs também possuem um Rank, escolhido pelo GM.

Quando você enfrenta alguém com um Rank maior, as Dificuldades de seus Desafios são aumentadas em 1 (mesmo se isso coloque-as acima de 6). Contra um adversário de Rank menor, elas são reduzidas em 1. Oponentes com 3+ Ranks acima ou abaixo, as regras de [Níveis de Poder](#Níveis-de-Poder) são usadas.

#### Derrotando Adversários
Como regra padrão, a maioria dos adversários requerem um número de pontos de Esforço igual ao seu Rank para serem derrotados. Um adversário "chefe" requer o dobro desse número e pode estar um Rank acima de seus parentes mais fracos.

### Magia & Psiônicos
Essas habilidades arcanas são representadas como Regalias.

Um personagem com tal Regalia pode fazer qualquer coisa que uma pessoa treinada poderia alcançar com ferramentas apropriadas, desde que se adéque tematicamente ao seu tipo de magia. Gastar Carma permite que o mago execute façanhas mágicas ainda maiores, superando quaisquer Desafios como um profissional poderia com equipamentos especializados.

A magia sempre deve ser pensada em termos de resultado. Não importa se o psiônico desintegra a porta ou simplesmente faz a fechadura se abrir – o objetivo é atravessar a porta.

Mas não se esqueça: a magia é, primariamente, uma ferramenta narrativa e não altera as mecânicas do jogo. Se o GM pedir um Desafio ágil para abrir a fechadura, o mago ainda deve *resolvê-lo* como um Desafio ágil, mesmo se utilizar magia para superar a porta.

#### Limitações Mágicas
Se a Regalia mágica for exageradamente ampla em escopo, o GM exigirá que o jogador selecione uma **Limitação**. Veja alguns exemplos abaixo:

- **Cansativa:** Seus feitiços lhe drenam e podem envolver sacríficos de sangue. Você gasta Determinação ao invés de Carma para usar magia.
- **Chamativa:** Você precisa fazer gestos e encantações para conjurar feitiços, tornando óbvio quando você está usando magia.
- **Destrutiva:** Seus feitiços causam muitos danos ambientais e colaterais.
- **Focada:** Você requer um foco para canalizar seus feitiços, como uma varinha, um cajado ou um símbolo sagrado. Se perdido ou quebrado, o item de fodo requer vários dias para ser substituído.
- **Material:** Sua magia requer uma fonte de energia ou matéria adequada e próxima para ser usada. Você não pode conjurar as coisas a partir do nada.
- **Pessoal:** Sua magia funciona em seu próprio corpo e posses, mas não pode ser usada diretamente em outros.
- **Ritualística:** Seus feitiços mais fortes requerem tempo para serem conjurados. Você não pode gastar qualquer Carma neles, exceto se tiver alguns minutos para se preparar.

#### Lista de Feitiços
Em alguns casos, os feiticeiros aprendem listas fixas de feitiços predefinidos. Para tal, mais alguns requerimentos são adicionados aos indivíduos que possuem uma Regalia arcana desse tipo.

Os magos devem escolher três feitiços durante a criação do personagem; estes determinam como e quando você pode usar sua magia.

Cada feitiço precisa ter um nome e uma limitação (e.g. <span class="spell">Raio Flamejante (destrutivo)</span>, <span class="spell">Escudo Relâmpago (pessoal)</span>, <span class="spell">Disfarce Ilusório (ritualístico)</span>, etc.).

Você pode criar feitiços mais potentes inserindo limitações adicionais. Isso deve restringir seu escopo e o GM será mais flexível na interpretação de seus efeitos. Talvez sua <span class="spell">Esfera de Invisibilidade</span> seja ritualística *e* requeira um foco, mas pode ser usada para esconder o grupo todo.

Novos feitiços são aprendidos durante o jogo, conforme critério do GM – talvez um mago possa aprender feitiços a partir de pergaminhos ou livros de feitiços ou talvez o GM recompense feiticeiros com um novo feitiço toda vez que ele adquirir um avanço.

### Montarias & Lacaios
Cavaleiros de armadura frequentemente montam seus corcéis leais em batalha, enquanto necromantes temíveis levantam lacaios mortos-vivos para servi-los em combate. Esses ajudantes geralmente podem ser abstraídos como Equipamentos – tratados como descrições de fundo, descrevendo suas ações como parte de suas rolagens de Desafio ou tratados como uma Regalia, caso você queira um benefício mecânico.

### Raças
A **Raça** de cada personagem deve ser determinada durante a criação do personagem.

Raças são tratadas como Regalias (talvez um elfo possa gastar um ponto de Carma para enxergar em escuridão total ou relembrar técnicas de batalha élficas anciãs) e como Peculiaridades (talvez um meio-orc seja tratado como um pária e possua pouco entendimento sobre etiqueta ou cultura humana), dependendo da situação e necessidade da história.

Durante a criação do personagem, informe os benefícios e malefícios de sua Raça.

### Níveis de Poder
Um semi-deus poderia facilmente subjugar um humano normal. De modo semelhante, um vampiro ou lobisomem dominaria um mero mortal, um piloto de *mech* superaria completamente um soldado de infantaria regular e um super-herói cósmico poderia obliterar um super-herói de bairro.

Na maioria dos casos, esses cenários são tratados usando interpretações relativas de sucesso e fracasso, atribuindo Aflições apropriadas ao personagem e à história. Um super-herói alienígena invulnerável poderia não ser ferido pelos disparos de uma arma de fogo, mas os ricochetes ainda poderiam matar os espectadores inocentes que ele jurou proteger, resultando na perda de Determinação. Ficar sem Determinação ainda levaria à derrota – talvez ele receba uma Aflição de <span class="affliction">Humilhação</span> ou <span class="affliction">Desespero</span>, conforme os gangster fogem do tiroteio, deixando o pobre super-herói para lidar com a imprensa furiosa e uma pilha de corpos.

Em outros casos, um adversário pode não representar ameaça nenhuma. Nesse caso, não há necessidade de rolar, apenas narrar o resultado.

Essas situações dependem da diferença de Rank entre os personagens, mas também do Conceito de cada um, conforme critério do GM.

### Cercos & Batalhas Épicas
Algumas vezes, o combate ocorre entre exércitos ao invés de indivíduos, cada lado mantendo seu próprio conjunto de pontos. O PC comandante utiliza Desafios astutos para controlar suas forças e eliminar todos os pontos do lado opostos, com o GM atribuindo a Dificuldade para atacar ou defender conforme o poder relativo e as vantagens táticas dos dois exércitos.

PCs individuais podem fazer uma diferença em tais conflitos, mas os riscos são grandes. Esses heróis podem eliminar pontos do exército oponente, mas se falharem em suas rolagens de Desafios defensivos, eles perdem seus *próprios* pontos de Determinação.

#### Batalhas Indiretas
A abordagem acima também é usada para outros cenários, como uma batalha legal entre duas empresas, guerras de territórios entre gangues ou guildas, guerras cibernéticas entre nações ou organizações de alta tecnologia, etc.

Como sempre, indivíduos podem contribuir seus esforços, mas os riscos serão mais altos.

### Super-Heróis & Super-Vilões
Se o seu personagem possui superpoderes e, considerando a temática central do jogo ele provavelmente terá, você deverá basear seu Conceito em sua personalidade mundana e lidar com seus superpoderes como Regalias. Portanto, você poderia ser um jornalista ágil com sua Regalia <span class="perk">Poderes de Aranha</span> ou um inventor podre de rico astuto com uma Regalia <span class="perk">Armadura de Ferro</span>. Como sempre, o GM dará mais impacto para Regalias definidas mais estritamente do que as definidas amplamente. Superpoderes exageradamente amplos também devem receber Limitações da mesma forma que magia (veja [Magia & Psiônicos](#Magia-&-Psiônicos)).

#### Limitações de Poderes
Assim como com magia, se um superpoder for exageradamente amplo em escopo, então você deve incluir uma Limitação. Veja alguns exemplos:

- **Aterrado:** A maioria dos super-humanos possui alguma forma de locomoção aérea, como voo, movimentação por teias, escorregadores de gelo, super-salto, etc. Mas você está limitado a usar meios de transporte mundanos.
- **Dispositivo:** Suas habilidades são concedidas por engenhocas ou equipamentos especiais, que podem ser temporariamente desabilitados ou perdidos.
- **Inofensivo:** Seus poderes não podem ser usados para atacar alguém diretamente.
- **Instável:** Você carece de controle total sobre seus poderes. Sempre que você tentar gastar Carma nessa Regalia, o GM pode introduzir uma Complicação ao invés de aceitar o Carma.
- **Intimidator:** Suas habilidades se manifestam de uma forma que invocam sensações de medo e pavor, aterrorizando os inocentes.
- **Negação:** Você não pode usar seus poderes enquanto estiver exposto à uma substância específica.
- **Vista-se:** Você deve mudar sua forma ou aparência para utilizar seus poderes.

### Sobrenaturais
Quando alguém é derrotado, ele adquire uma Aflição. Caso um PC seja derrotado pela mordida ou garras infecciosas de uma criatura sobrenatural, o GM pode atribuir uma Aflição apropriada, como <span class="aff">Licantropia</span>, <span class="affliction">Vírus Zumbi</span>, <span class="affliction">Vampirismo</span>, etc.

Ao contrário das Peculiaridades, que são ativadas pelo jogador, o GM decide quando e como as Aflições são usadas durante o jogo. Um lobisomem recém-infectado não possui controle sobre sua transformação ou da carnificina  que eles causam – mas ele pode, posteriormente, converter sua Aflição em uma Peculiaridade, conforme aprende a controlar sua condição.

Da mesma forma, Aflições não fornecem benefícios, mas um jogador pode, posteriormente, adquirir Regalias Sobrenaturais como <span class="perk">Força Sobrenatural</span> ou <span class="perk">Garras Rasgantes</span>. Se o PC selecionar uma Regalia ampla que englobe uma variedade de habilidade, ele também deve selecionar uma Limitação (da mesma forma que a Limitação Mágica) – por exemplo, a Regalia <span class="perk">Dons do Lobisomem</span> só poderia ser aplicada se o personagem se transformam em sua forma lupina.

#### Removendo Aflições
Se uma Aflição não for permanente, o GM pode oferecer uma forma baseada na história para removê-la – talvez matando o vampiro que te mordeu reverteria sua condição ou existe uma cura para o vírus zumbi. Aflições permanentes também podem ser removidas assim, mas isso custa Carma permanentemente.

Soluções mais drásticas também podem ser permitidas, como converter uma Aflição <span class="affliction">Vírus Zumbi</span> em uma Peculiaridade <span class="quirk">Perna Amputada</span> (com o custo de avanço usual).

#### Declínio Gradual
Algumas condições *não* oferecem quaisquer benefícios. Talvez sua infecção te transforme em uma fera sem mente movida pela fúria, instinto e fome.

Tais infecções podem assumir a forma de um lento declínio – vítimas de uma mordida zumbi poderiam sobreviver horas ou até mesmo dias antes de eventualmente sucumbirem. O GM pode até mesmo usar Aflições futuras para representar a transformação gradual de um personagem.

### Veículos
Quer você seja um capitão pirata, um motorista de tanque, um comandante de nave-espacial, um motoqueiro fantasma ou um piloto de *mech*, todos os veículos são tratados de maneira semelhante.

Eles são tratados como Equipamentos mundanos ou como Regalias se forem particularmente poderosos. Mas, se o veículo for excepcionalmente poderoso ou importante, ele pode ser construído como um personagem.

#### Veículos como Personagens
Veículos importantes iniciam com um Conceito (mas não um Traço), uma Regalia e uma Peculiaridade. Eles não possuem Carma, mas começam com três pontos de Determinação e podem avançar, conforme critério do GM.

O motorista (ou piloto) realiza rolagens de Desafio usando seu próprio Traço, mas pode utilizar o Conceito, Regalias e Peculiaridades de seu veículo como se fossem seus.

Em combate entre veículos de força significativamente diferente – como um caça espacial contra um couraçado – as regras de Níveis de Poder são usadas.
